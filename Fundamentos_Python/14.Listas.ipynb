```python
#Quero saber a soma de todas as idades
idades = [28, 42, 43, 35, 39, 28, 38]
print('Soma das idades', sum(idades))             
print('Media das idades', sum(idades)/len(idades)) 
print('Menor idade', min(idades))
print('maior idade', max(idades))
```
