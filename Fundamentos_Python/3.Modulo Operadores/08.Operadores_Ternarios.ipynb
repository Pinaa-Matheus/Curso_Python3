Operadore Ternarios

```python
esta_chuvendo = False
# A variável booleana 'esta_chuvendo' é usada para indexar o set.
# False é interpretado como 0 e True como 1, escolhendo assim o estado da roupa.
# ATENÇÃO: Embora funcione, sets não garantem ordem, tornando esta prática arriscada.
'Hoje estou com as roupas ' + {'secas.', 'molhadas.'} [esta_chuvendo]
```

Modo mais Claro de fazer

```python
esta_chuvendo = True
'Hoje estou com as roupas ' + ('Molhadas.' if esta_chuvendo else 'secas.')
# Hoje estou com as roupas molhadas SE ESTIVER CHOVENDO se NAO ESTIVER chovendo elas estao secas
```
