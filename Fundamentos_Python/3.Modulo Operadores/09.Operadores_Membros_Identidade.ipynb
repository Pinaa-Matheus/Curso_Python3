Operadores de Membro

```python
lista = [1, 2, 3, 'Ana', 'Carla']

print(2 in lista)           #Dois esta na lista 
print('Ana' not in lista)   #Ana nao esta na lista
```

Operadores de Identidade

```python
x = 3 
y = x 
z = 3 

print(x is y)
print(y is z)
print(x is not z)
```

Exemplo

```python
#Sera que os operadores de identidade terao a mesma logica com listas

lista_a = [1, 2, 3]
lista_b = lista_a
lista_c = [1, 2, 3]

print(lista_a is lista_b)
print(lista_b is lista_c)
#Na teoria era para dar true porque eles tem os mesmos valores, entao porque nao deu True?
#Nessa caso existem duas lista na memoria a lista_a e a lista_c elas parecem iguais porem sao diferentes 
#Como nao estamos trabalhando com valores a logica e diferente entao para resumir sao duas lista 'iguais' porem direrentes
print(lista_a is not lista_c)
```
