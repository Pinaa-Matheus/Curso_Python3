Operadores Logicos sao usados para modificar expressoes booleanas(verdadeiro/falso)

```python
True or False
7 != 3 and 2 > 3 ##!= Diferença 
```

Tabela And 

(Vai retonar verdadeiro APENAS se as duas condicoes serem verdadeiras)

```python
#Tabela verdade do AND
True and True
True and False
False and True
False and False

#As duas condições que voce escreveu tem que ser atingida
```

Tabela OR 

(Vai retonar verdadeiro SE PELO MENOS uma das declaracoes forem VERDADEIRAS, caso ao contrario FALSO)

```python
#Tabela verdade do OR
True or True
True or False
False or True
False or False

#Basta ter somente um verdadeiro 
```

Operador de negação (Unário)

(NOT inverte a condicao booleana, ou seja: se A e verdadeiro, not A retorna false; se B e falso, not B retorna verdadeiro)

```python
not True
not True

#Ele so nega 

not 0 # 0 = False 
not 1 # 1 = True (Qualquer numero diferente(!=) de 0 sempre e true)

not not -1
# Nessa operacao ele faz primeiro o not -1 ai depois com o resultado do not -1 ele aplica o not no resultado do not -1 
```

Tabela do XOR

```python
True != True
True != False
False != True
False != False

#Tem que Um ou outro diferente 
```

```python
#CUIDADO! (Operações BIT A BIT)
True & False
False | True
True ^ False

#   3       = 11
#   2       = 10
# Resultado = 10
3 & 2  
```
