Dicionarios

```python
#Dicionarios sao pares de chaves/valor 

dados_mat = {
    'sobrenome': 'Pina',    #Nao importa o valor
    'nome': 'Matheus',  
    'Filhos':True,
    'formacao':['engenharia da computacao', 'informatica'],
    'cargos':[
        {'nome': 'Ensino medio', 'empresa': 'Msul'},
        {'nome': 'Ensino medio', 'empresa': 'Senac'},
        {'nome': 'Graduacao', 'empresa': 'Senac'},
    ]
    }
#Sempre sera primeiro a ch ave depois valor
print(dados_mat)
```

Acessando as informacoes  

```python
#Acessando valore de um dicionarios

#dados_mat[0] #Nao funciona que nem as listas 
dados_mat['nome'] #O que muda e que a gente puxa pela chave  
print(dados_mat['formacao'][-1])
print(dados_mat['cargos'][0]['empresa'])
```

```python
dados_mat['cargos'][-1]
```

Adicionar novos dados dentro do dicionario 

```python
dados_mat['estado_civil'] = 'Solteiro'
print(dados_mat)
```

Descobrir as chaves de um dicionario

```python
dados_mat.keys()
```

Valores

```python
dados_mat.values()
#Aqui sao so sao os valores 
```

Items (Porem colocadas em tuplas e na ordem de chave e valor)

```python
dados_mat.items()
```

"Andandado" pelo dicionario

```python
for i in dados_mat:
    print(i, dados_mat[i])
```

```python
for item in dados_mat.items():
    print(item)
```

```python
#unpacking
for chave, valor in dados_mat.items():
    print(chave, '->' ,valor)
```
