```python
nome = 'Ana Paula'
nome[0]  #A
nome[6]  #U
nome[-3] #U (Conseguimos ler tanto da esquerda para direita para direita para esquerda)
nome[4:] #Range (Aqui ele comeca a pegar da posição 4 ate terminar a string)
nome[-5:]#Range
nome[:3]
#[Inicio:Fim]
```

```python
numeros = '1234567890'
numeros
numeros[::]
numeros[::2] #Aqui seria os steps ou como a gente escreveu pegando em dois em dois
#[Inicio:Fim:Steps]
numeros[::-1] #Invertendo uma String
```
