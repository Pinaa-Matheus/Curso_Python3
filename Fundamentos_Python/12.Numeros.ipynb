Numero Decimais

```python
1.1 + 2.2
#Os computadores não entendem números decimais da mesma forma que nós.
```

```python
from decimal import Decimal, getcontext
Decimal(1) / Decimal(7) #Numero gigante

getcontext().prec = 4 #Aqui estamos colocando um contexto na nossa operacao de quatro casas decimais
Decimal(1) / Decimal(7)

Decimal.max(Decimal(1), Decimal(7)) #Maior Numero
```
